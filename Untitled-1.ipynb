```python
%pip install seaborn scikit-learn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

# 1. LOAD THE DATASET
print("🔄 Step 1: Ingesting data...")
df = pd.read_csv("IMDB Dataset.csv")

# Quick text cleaning baseline to ensure uniform strings
df["review"] = df["review"].str.lower().str.replace(r"<br\s*/?>", " ", regex=True)

# 2. SEPARATE INTO TRAIN AND TEST SETS (CROSS-VALIDATION HOLDOUT)
print("✂️ Step 2: Splitting data into 80% Training and 20% Testing sets...")
X_train, X_test, y_train, y_test = train_test_split(
    df["review"], df["sentiment"], test_size=0.20, random_state=42
)

# 3. FEATURE EXTRACTION (TEXT TO NUMBERS VIA TF-IDF)
print("🧮 Step 3: Transforming text into mathematical matrices via TF-IDF...")
# We limit features to top 10,000 words to optimize memory safety
tfidf = TfidfVectorizer(max_features=10000, stop_words="english")
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# 4. ALGORITHM SELECTION & TRAINING (LOGISTIC REGRESSION)
print("🧠 Step 4: Training the Logistic Regression classifier model...")
model = LogisticRegression(max_iter=1000)
model.fit(X_train_tfidf, y_train)
print("✅ Training complete!")

# 5. MODEL EVALUATION & METRICS GENERATION
print("\n📊 Step 5: Generating Performance Metrics Report...")
y_pred = model.predict(X_test_tfidf)

# Print metrics to screen for your report
print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred))

# Generate Confusion Matrix Chart
print("\n🎨 Creating Confusion Matrix visual chart...")
cm = confusion_matrix(y_test, y_pred, labels=["negative", "positive"])
plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Purples",
    xticklabels=["Negative", "Positive"],
    yticklabels=["Negative", "Positive"],
)
plt.title("Confusion Matrix: Sentiment Classifier")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.tight_layout()
plt.savefig("sentiment_confusion_matrix.png")
plt.close()
print("🎉 Success! Check your folder for 'sentiment_confusion_matrix.png'!")
```

```python
import sys
import matplotlib.pyplot as plt
import nltk
import pandas as pd
import seaborn as sns
import spacy
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Append your explicit local user package directory to the path registry
sys.path.append(r"C:\Users\Sravya\AppData\Roaming\Python\Python314\site-packages")

# Download VADER lexicon
nltk.download("vader_lexicon")

# 1. LOAD DATASET AND SAMPLE FOR COMPUTATIONAL EFFICIENCY
print("🔄 Step 1: Loading dataset and sampling records...")
df = pd.read_csv("IMDB Dataset.csv")
df_sample = df.sample(n=500, random_state=42).copy()

# 2. SENTIMENT ANALYSIS BLOCK (LEXICON-BASED VADER)
print("🧠 Step 2: Running VADER Sentiment Analyzer...")
sia = SentimentIntensityAnalyzer()


def analyze_vader_sentiment(text):
    score = sia.polarity_scores(text)["compound"]
    return "positive" if score >= 0.05 else "negative"


df_sample["predicted_sentiment"] = df_sample["review"].apply(
    analyze_vader_sentiment
)

correct_predictions = (
    df_sample["predicted_sentiment"] == df_sample["sentiment"]
).sum()
vader_accuracy = correct_predictions / len(df_sample)
print(f"✅ VADER Sentiment Identification Accuracy: {vader_accuracy:.4f}")

# 3. NAMED ENTITY RECOGNITION BLOCK (spaCy Pipeline)
print("🔍 Step 3: Initiating spaCy Named Entity Recognition...")

# Force load the model using its explicit module path shortcut string
import en_core_web_sm

nlp = en_core_web_sm.load()

entity_counts = {"PERSON": 0, "ORG": 0, "GPE": 0}
entity_sentiment_map = []

# Process text fields to extract entities
for idx, row in df_sample.iterrows():
    doc = nlp(row["review"][:1000])
    for ent in doc.ents:
        if ent.label_ in entity_counts:
            entity_counts[ent.label_] += 1
            entity_sentiment_map.append(
                {"Entity": ent.text, "Type": ent.label_, "Sentiment": row["sentiment"]}
            )

df_entities = pd.DataFrame(entity_sentiment_map)
print("\n=== TOP ENTITY TYPES FOUND ===")
print(df_entities["Type"].value_counts())

# 4. DATA VISUALIZATION GENERATION
print("\n🎨 Creating Entity-Sentiment correlation visualization...")
plt.figure(figsize=(8, 5))
sns.countplot(
    data=df_entities,
    x="Type",
    hue="Sentiment",
    palette=["#800080", "#4B0082"],
    edgecolor="black",
)
plt.title("Named Entity Extractions Categorized by Surrounding Sentiment")
plt.xlabel("Entity Category Type")
plt.ylabel("Extraction Volumetric Counts")
plt.tight_layout()
plt.savefig("ner_sentiment_distribution.png")
plt.close()
print("🎉 Success! Check your folder for 'ner_sentiment_distribution.png'!")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# 1. INGEST DATASET AND CREATE DOCUMENT LENGTH QUANTILES (TIME/SCALE PROXY)
print("🔄 Step 1: Ingesting dataset and engineering text scale partitions...")
df = pd.read_csv("IMDB Dataset.csv")

# Clean baseline strings to keep execution smooth
df["clean_review"] = df["review"].str.replace(
    r"<br\s*/?>", " ", regex=True
).str.lower()
df["word_count"] = df["clean_review"].apply(lambda x: len(x.split()))

# Partition reviews into 3 structural tiers representing the evolution of narrative depth
df["narrative_tier"] = pd.qcut(
    df["word_count"],
    q=3,
    labels=["Short (Concise)", "Medium (Standard)", "Long (In-Depth)"],
)

# 2. FEATURE EXTRACTION FOR TOPIC MODELING
print("🧠 Step 2: Extracting feature tokens via CountVectorizer...")
# Using 3,000 top words and filtering standard stop words
vectorizer = CountVectorizer(
    max_features=3000, stop_words="english", max_df=0.95, min_df=2
)
tf_matrix = vectorizer.fit_transform(df["clean_review"])
feature_names = vectorizer.get_feature_names_out()

# 3. ADVANCED TOPIC MODELING VIA LDA (LATENT DIRICHLET ALLOCATION)
print("🔍 Step 3: Optimizing Latent Dirichlet Allocation (LDA) for 3 topics...")
lda = LatentDirichletAllocation(
    n_components=3, random_state=42, max_iter=5, learning_method="online"
)
lda_topics = lda.fit_transform(tf_matrix)

# Map extracted topics back to names based on dominant words
print("\n=== TOP LOG-LIKELIHOOD WORDS PER EXTRACTED TOPIC ===")
for topic_idx, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[: -6 : -1]]
    print(f"Topic #{topic_idx + 1}: {', '.join(top_words)}")

# Assign the most dominant topic to each review record
df["dominant_topic"] = np.argmax(lda_topics, axis=1) + 1

# 4. QUANTIFY LINGUISTIC TREND EVOLUTION OVER NARRATIVE TIERS
print("\n📈 Step 4: Measuring evolution of language trends over narrative scales...")
trend_matrix = (
    pd.crosstab(
        df["narrative_tier"],
        df["dominant_topic"],
        normalize="index",
    )
    * 100
)

# Rename topics for clear reporting presentation
trend_matrix.columns = [
    "Topic 1: Production & Plot",
    "Topic 2: Emotional Response",
    "Topic 3: Genre & Character Analysis",
]
print(trend_matrix)

# 5. GENERATE FINAL PRODUCTION VISUALIZATION
print("\n🎨 Generating language trend evolution chart...")
plt.figure(figsize=(10, 6))
trend_matrix.plot(
    kind="bar",
    stacked=True,
    color=["#4B0082", "#800080", "#D8BFD8"],
    edgecolor="black",
    ax=plt.gca(),
)

plt.title("Evolution of Linguistic Topic Trends Across Narrative Scales", fontsize=14)
plt.xlabel("Review Narrative Depth (Word Count Tiers)", fontsize=12)
plt.ylabel("Topic Distribution Density Percentage (%)", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Linguistic Themes (LDA Topics)", loc="lower right")
plt.tight_layout()

# Save final asset
plt.savefig("linguistic_trend_evolution.png")
plt.close()
print("🎉 Success! Your capstone visualization 'linguistic_trend_evolution.png' is saved!")
```
